```python
import pandas as pd
```

```python
import random
import numpy as np
```

```python
lst = ['robot'] * 10 
lst += ['human'] * 10 
random.shuffle(lst) 
df = pd.DataFrame({'whoAmI':lst})
data = df.copy()
data
```

```python
data.head
```

```python
# То что мы должны получить после применения своего кода
data1 = pd.get_dummies(data['whoAmI'])
data1.head()
```

# Вариант 1

```python
data.loc[data['whoAmI'] == 'robot', 'iAmRobot'] = 1
data.loc[data['whoAmI'] == 'human', 'iAmRobot'] = 0
data.loc[data['whoAmI'] == 'robot', 'iAmHuman'] = 0
data.loc[data['whoAmI'] == 'human', 'iAmHuman'] = 1
data2 = pd.DataFrame(data[['iAmRobot', 'iAmHuman']])
data2
# тут я не могу понять почему значения выводятся как float
```

# Вариант 2

```python
data3 = df.copy()
data3['iAmRobot'] = data3['whoAmI'].apply(lambda x: 1 if x == 'robot' else 0)
data3['iAmHuman'] = data3['whoAmI'].map({'human': 1, 'robot': 0})
data3 = data3.drop('whoAmI', axis=1)
data3.head()
```

# Вариант 3

```python
data4 = df.copy()
data4['iAmRobot'] = np.where(data4['whoAmI'] == 'robot', 1, 0)
data4['iAmHuman'] = (data4['whoAmI'] == 'human').astype(int)
data4 = data4.drop('whoAmI', axis=1)
data4.head()
```
